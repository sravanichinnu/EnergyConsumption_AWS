# Feature Engineering for Time-Series Forecasting

This notebook explains the feature engineering strategy used for forecasting PJME hourly energy demand.

```python
import pandas as pd

df = pd.read_parquet(
    "s3://energy-consumption-forecasting-project/processed/pandas/pjme_energy_features.paruet"
)

feature_cols = [
    "lag_1", "lag_24", "lag_168",
    "rolling_mean_24", "rolling_std_24",
    "hour", "day_of_week", "month"
]

df[feature_cols].describe()
```

- Lag features capture short-term and weekly seasonality.
- Rolling statistics smooth short-term fluctuations.
- Calendar features provide contextual temporal signals.
